#### Question1

Write a function using only list filter lambda that can tell whether a number is a Fibonacci number or not. You can use a pre-calculated list/dict to store fab numbers till 10000

```
### Function to calculate the n-th Fibonacci number in the series
def fib(n):
    if n < 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

# Storing all the Fibonacci number till 25th number in the series
fib_store = [fib(x) for x in range(25)]

# Maximum number in the fib_store
print(f'Total numbers in fib_store: {len(fib_store)} \nMaximum number in the fib_store: {max(fib_store)}')
```

```
# Function to check whether 
# def check_fib(x):
#     return x in fib_store

check_fib = lambda x: x in fib_store

# Checking the function for Fibonacci
print(check_fib(144), check_fib(14), check_fib(0), check_fib(5), check_fib(1), check_fib(100), check_fib(377))
```

```
# List of even numbers till 9999
l1 = [ x for x in range(10000) if x%2 == 0]

# Check all the fibonacci number in the even number set 0 - 10_000
list(filter(lambda x: check_fib(x), l1))
```

```
# List of odd numbers till 9999
l1 = [ x for x in range(10000) if x%2 == 1]

# Check all the fibonacci number in the list l1
list(filter(lambda x: check_fib(x), l1))
```

### Question 4

Using reduce functions:

1. add only even numbers in a list
2. find the biggest character in a string (printable ascii characters)
3. adds every 3rd number in a list

```
# Ques 4.1: add only even numbers in a list

import random
n = 5
l_rand = random.sample(range(25), n)

from functools import reduce

print(f'The random input list of length {n}: {l_rand}')

# The output from method1:
result1 = reduce(lambda a,b: a+b, [x for x in l_rand if x%2==0])
print(f'The output from method1: {result1}')

# The output from method2:
result2 = reduce(lambda a,b: a+b, filter(lambda x: x%2 == 0, l_rand))
print(f'The output from method2: {result2}')
```

```
# Ques 4.2: Find the biggest character in a string (printable ascii characters)

#Input String
l_str = '<\kilimanjaro < K < HIMALAYA \>'

# method1
max_char = reduce(lambda a,b:  max(a,b)  ,[ord(char) for char in l_str])
print(f'The maximum charcter in the string <method1>: `{chr(max_char)}` and the ascii value: `{max_char}`')


# method2
max_char = max(list(map(lambda x: ord(x),[char for char in l_str])))
print(f'The maximum charcter in the string <method2>: `{chr(max_char)}` and the ascii value: `{max_char}`')
```

```
# Ques 4.3: Adds every 3rd number in a list

import random
from functools import reduce

n = 13
l_rand = random.sample(range(25), n)

print(f'The random input list of length {n}: {l_rand}')

# The output from method1:
result1 = reduce(lambda a,b: a+b, [x for x in l_rand if l_rand.index(x)%3 == 2])
print(f'The output from method1: {result1}')


# The output from method2:
result2 = reduce(lambda a,b: a+b, [x for x in l_rand if l_rand.index(x)%3 == 2])
print(f'The output from method1: {result1}')
```

### Question5:

Using randint, random.choice and list comprehensions, write an expression that generates 15 random KADDAADDDD number plates, where KA is fixed, D stands for a digit, and A stands for Capital alphabets. 10<<DD<<99 & 1000<<DDDD<<9999 PTS:100

```
import random
import string


num = 15
num_plates = [''.join(['KA', str(random.randint(10, 99)), \
                ''.join(random.choices(string.ascii_uppercase, k=2)), \
                        str(random.randint(1000, 9999))]) for i in range(num)]


print(f'The Number plates with all requirements: \n{num_plates}')
```
